## 3. Какая средняя площадь квартир и медиана. Какой процент выбросов?

```python
square = df["Площадь, м2"]
```

```python
ax = sns.displot(square, kde=True)
ax.set(ylabel="Количество квартир")
plt.axvline(x=square.median(),
            color='red',
            ls='--', 
            lw=1)
plt.axvline(x=square.mean(),
            color='green',
            ls='--', 
            lw=1)
```

```python
ax = sns.boxplot(x=square.values)
ax.set(xlabel="Площадь, м2")
```

```python
square.describe()
```

```python
# Посчитаем процент выбросов
def quantity_and_percentage_outliers(x: pd.Series) -> str:
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)

    outliers = x[(x < lower_bound) | (x > upper_bound)]
    percent_outliers = outliers.count() / x.count() * 100
    print(f"Выбросы: {outliers.count()} или {round(percent_outliers)}%.")
    
quantity_and_percentage_outliers(square)
```

- Средняя площадь 3-х комнатных квартир составляет 68.5 м2, а медиана 63 м2. 
- У 113 квартир или 8%, площадь значительно отличается от площади остальных квартир.

Изучим подробнее отношение цены и остальных количественных переменных. Для этого произведем логарифмирование наших количественных переменных, это позволит выровнять масштаб данных, улучшить линейность отношений, стабилизировать экстремальные значения переменных. Также проверим распределения на нормальность и выберем какой коэффицент корреляции использовать.

```python
# Логарифмируем переменные
df_log = df[['Площадь, м2', 'Этаж', 'Всего этажей', 'Цена, ₽', 'Цена за квадрат, ₽/м²']].apply(np.log)
df_log.head()
```

```python
# Для проверки на нормальность используем тест Шапиро–Уилка.
def test_shapiro_wilk(df: pd.DataFrame | pd.Series, probability_level: float=0.05) -> None:
    if type(df) == pd.DataFrame:
        columns = df.columns
        total_res = []
        for column in columns:
            shapiro_test = stats.shapiro(df[column])
            if shapiro_test[1] > probability_level:
                res = f"- True, {df[column].name}: p>{probability_level}, нулевая гипотеза о нормальности распределения не отвергается"
                total_res.append(res)
            else:
                res = f"- False, {df[column].name}: p<{probability_level}, нулевая гипотеза о нормальности распределения отвергается, принимается альтернативная гипотеза о том, что данные не имеют нормальное распределение"
                total_res.append(res)
        for res in total_res:
            print(res)
    else:
        res = ""
        shapiro_test = stats.shapiro(df)
        if shapiro_test[1] > probability_level:
            res = f"- True, {df.name}: p>{probability_level}, нулевая гипотеза о нормальности распределения не отвергается"
        else:
            res = f"- False, {df.name}: p<{probability_level}, нулевая гипотеза о нормальности распределения отвергается, принимается альтернативная гипотеза о том, что данные не имеют нормальное распределение"
        print(res)

test_shapiro_wilk(df_log)
```

## 9. Сформулиовать и проверить гипотезу о средней цене на 3-х комнатные квартиры

- Генеральная совокупность:\
    Объявления с сайта Циан о 3-х комнатных квартирах, во вторичном и многоэтажном жилье, за октябрь 2023 года, в городе Омск. 

- Гипотезы:\
    H0 - средняя цена на 3-х комнатые квартиры, в городе Омск, составляет 6 550 000 ₽\
    H1 - средняя цена не равна 6 550 000 ₽

Для проверки гипотезы воспользуемся центральной предельной теоремой, построим доверительные интервалы, проведем одновыборочный Т-тест.

Центральная предельная теорема гласит, что если мы будем много раз выборочно собирать данные, то среднее средних всех выборок будет стремится к среднему генеральной совокупности.

```python
price = df["Цена, ₽"]
price.describe()
```

```python
# Распределение генеральной совокупности
ax = sns.displot(price, kde=True)
ax.set(ylabel="Количество квартир")
plt.axvline(x=price.median(),
            color='red',
            ls='--', 
            lw=1)
plt.axvline(x=price.mean(),
            color='green',
            ls='--', 
            lw=1)
```

```python
# Возьмем 50 выборок, каждая размером 200.
# Соберем значения средних наших выборок в 1 список
# Посчитаем доверительные интервалы для каждой выборки и построим график с доверительными интервалами
sample_mean = np.zeros(50)
for i in range(50):
    sample = price.sample(200)
    mean = sample.mean()
    # standard error of the mean
    sem = stats.sem(sample)
    sample_mean[i] = mean
    # confidence interval
    ci = stats.norm.interval(confidence=0.95, loc=mean, scale=sem)
    plt.errorbar(x=mean, y=i, xerr=ci[1]-mean, fmt='o')

plt.axvline(x=sample_mean.mean(),
            color='green',
            ls='--', 
            lw=1)
plt.title("Доверительные интервалы для среднего")
```

95% наших доверительных интервалов будут включать истинное среднее генеральной совокупности

```python
# Средние наших выборок должны быть распределены нормально, проверим это
sample_mean = pd.Series(sample_mean,name="distribution_sample_mean")
sample_mean.describe()
```

```python
# Построим распределение выборочных средних
ax = sns.displot(sample_mean, kde=True)
ax.set(ylabel="Количество средних по выборке")
plt.axvline(x=sample_mean.median(),
            color='red',
            ls='--', 
            lw=1)
plt.axvline(x=sample_mean.mean(),
            color='green',
            ls='--', 
            lw=1)
```

```python
# Для проверки на нормальность используем тест Шапиро–Уилка.
test_shapiro_wilk(sample_mean)
```

```python
# Найдем доверительный интервал для выборочных средних
mean = sample_mean.mean()
sem = stats.sem(sample_mean)
ci = stats.norm.interval(confidence=0.95, loc=mean, scale=sem)
ci
```

Доверительный интервал: [6 492 903, 6 606 373] 

- В 95% случаев наш доверительный интервал будет включать истинное среднее генеральной совокупности.

Для проверки нулевой гипотезы, о том что средняя цена на 3-х комнатые квартиры, в городе Омск, составляет 6 550 000 ₽, проведем одновыборочный T-тест.

```python
# Одновыборочный T-тест
t_statistic, p_value = stats.ttest_1samp(sample_mean, 6_500_000)

# пороговым значением мы выбрали 0.05 или 5%
if p_value < 0.05:    
    print(f"{round(p_value,3)} < 0.05, есть основания отвергнуть нулевую гипотезу")
else:
    print(f"{round(p_value,3)} > 0.05, нет оснований отвергнуть нулевую гипотезу")
```

- Вывод:
    - Составлена нулевая и альтернативная гипотеза
        - H0 - средняя цена на 3-х комнатые квартиры, в городе Омск, составляет 6 550 000 ₽
        - H1 - средняя цена не равна 6 550 000 ₽
    - Для проверки гипотезы была использована центральная предельная теорема, найден доверительный интервал выборочных средних, проведен одновыборочный T-тест
    - Доверительный интервал: [6 492 903, 6 606 373]. В 95% случаев наш доверительный интервал будет включать истинное среднее генеральной совокупности.
    - Одновыборочный T-тест показал, что нет оснований отвергнуть нулевую гипотезу
